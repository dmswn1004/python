```python
# 응용2
# 구구단을 2차원 배열 형태로 출력
# 반복문 활용
gugu = []
for i in range(1,10):
    for j in range(1,10):
        gugu.append([i,j,i*j])
print(gugu)
```

```python
gugu = [[i,j,i*j] for i in range(1,10) for j in range(1,10)]
print(gugu)
```

```python
# 응용3
# 구구단 9~1단 순으로 출력
# 반복문 활용
gugu = []
for i in range(9,0,-1):
    for j in range(1,10):
        gugu.append([i,j,i*j])
print(gugu)
```

```python
gugu = [[i,j,i*j] for i in range(9,0,-1) for j in range(1,10)]
print(gugu)
```

```python
# 응용4
# 구구단 2,5단 제외하고 출력
# 반복문 활용
gugu = []
for i in range(1,10):
    for j in range(1,10):
        if i!=2 & i!=5:
            gugu.append([i,j,i*j])
print(gugu)
```

```python
gugu = [[i,j,i*j] for i in range(1,10) for j in range(1,10) if (i!=2)&(i!=5)]
print(gugu)
```

```python
# 응용5
# 두 수의 공약수 출력
# append 메소드 활용
a = 12
b = 6
cd = []
for i in range(1,b+1):
    if (a%i==0)and(b%i==0):
        cd.append(i)
print(cd)
```

```python
cd = [i for i in range(1,b+1) if (a%i==0) and (b%i==0)]
print(cd)
```

```python
# 응용 6
# 두 수의 최대 공약수
# append 메소드 활용
a = 12
b = 6
cd = []
for i in range(1,b+1):
    if (a%i==0)and(b%i==0):
        cd.append(i)
print(cd[-1])
```

```python
cd = [i for i in range(1,b+1) if (a%i==0) and (b%i==0)]
print(cd[-1])

gcd = max([i for i in range(1,b+1) if (a%i==0) and (b%i==0)])
print(gcd)
```

```python
# 응용7
# 두 수의 최소공배수
# append 메소드 활용
for i in range(a,a*b+1):
    if (i%a==0)and(i%b==0):
        print(i)
        break
```

```python
cm = [i for i in range(a,a*b+1) if (i%a==0)and(i%b==0)]
print(cm[0])

lcm = min([i for i in range(a,a*b+1) if (i%a==0)and(i%b==0)])
print(lcm)
```

```python
# 응용8
# 소수 아닌수 번호와 함께 출력
# append 메소드 활용
non_prime = []
for i in range(2,101):
    count = 0
    for j in range(1,i+1):
        if i%j==0:
            count += 1
    if count != 2:
        non_prime.append(i)
        
for i in range(len(non_prime)):
    print('%d: %d'%(i+1,non_prime[i]))
```

```python
non_prime = list(set([i for i in range(2,101) for j in range(2,i) if i%j==0]))
for i,j in enumerate(non_prime):
    print('%d: %d'%(i+1,j))
```
