```python
# [기본 1] 
# 1~100 홀수 만 출력
for i in range(1,101):
    if i%2==1:
        print(i)
# if i%2 != 0 가능
```

```python
# [기본 2] 1~9단 구구단 출력
for i in range(1,10):
    for j in range(1,10):
        print('%d * %d = %d'%(i,j,i*j))
    print(' ')
```

```python
# [기본 3] 9~1단 구구단 출력
for i in range(9,0,-1):
    for j in range(1,10):
        print('%d * %d = %d'%(i,j,i*j))
```

```python
# [기본 4] 2,5
for i in range(1,10):
    if (i!=2)&(i!=5):
        for j in range(1,10):
            print('%d * %d = %d'%(i,j,i*j))
```

```python
# [기본 5] *출력
for i in range(1,6):
    for j in range(1,(i+1)):
        print('*',end = '')
    print(' ')
```

```python
# 방법 2
for i in range(1,6):
    print('*'*i)
```
