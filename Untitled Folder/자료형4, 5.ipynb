```python
# 파이썬 프로그래밍 자료형4
# 군집 자료형
# 문자열
a='hello world'
print(a)
print(type(a))

a="hello world"
print(a)
print(type(a))

a='''
hello 
world
'''
print(a)
print(type(a))

a="""
hello 
world
"""
print(a)
print(type(a))

a="hello\nworld"
print(a)

# 인덱싱
print(a[0])
print(a[2])
print(a[-1])
print(a[-3])

# 슬라이싱
print(a[0:5])
print(a[0:6])
print(a[:5])
print(a[6:])
print(a[:])

```

```python
# 리스트
a=[1, 2, 3, 4, 5]
print(a)
print(type(a))

# 인덱싱
print(a[0])
print(a[1])
print(a[-1])

# 슬라이싱
print(a[1:4])

# 메소드
# append
a=[1, 2, 3]
print(a)
a.append(4)
print(a)

#insert
a.insert(4,5)
print(a)
a.insert(0,0)
print(a)

# pop
print(a.pop())
print(a)

# 리스트 속의 리스트
a = [['a','b','c'],1,2,3]
print(a)
print(a[0][0])
print(a[0][1])

help(list)
```

```python
# 튜플
a = (1,2,3)
print(a)
print(type(a))

print(a[0])
print(a[2])
print(a[:3])

# a[0] = 5 오류
```

```python
# 집합
a = {1,2,3}
print(a)
print(type(a))

# 집합은 순서가 없음
a = {3,1,2}
print(a)

# 중복 불가
a = {1,2,3,1,2,}
print(a)

# 합집합
a = {1,2,3}
b = {3,4,5}
print(a|b)

# 교집합
print(a&b)

# 차집합
print(a-b)
```

```python
# 딕셔너리
a = {'key1':'a','key2':'b','key3':3}
print(a)
print(type(a))
print(a['key3'])

a['key4'] = 'c'
print(a)

print(a.keys())
print(a.values())
print(a.items())
```

```python
# 파이썬 자료형 변환
# 단일 자료형 변환
# 정수형 변환
a = 1.0
print(type(a))
print(type(int(a)))
print(type(a))

a = int(a)
print(type(a))

# 실수형 변환
a = 1
print(type(a))
print(a)

a = float(a)
print(type(a))
print(a)

# 복소수의 변환
a = 1
print(type(a))

a = complex(a)
print(a)
print(type(a))

# 논리형의 변환
a = 1
print(type(a))
a = bool(a)
print(a)
print(type(a))
```

```python
# 군집 자료형 변환
# 문자열
a = 1
a = str(a)
print(a)
print(type(a))

# 리스트
a = 'hello'
print(a)
a = list(a)
print(a)
print(type(a))

# 튜플
a = [1,2,3]
a = tuple(a)
print(a)
print(type(a))

# 집합 변환
a = 'hello'
a = set(a)
print(a)
print(type(a))

# 딕셔너리 변환
a = [1,2,3]
# a = dict(a)
a = [[1,2],[3,4]]
a = dict(a)
print(a)
print(type(a))
```
