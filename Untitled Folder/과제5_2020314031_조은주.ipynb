```python
# 과제 5-1
# 방법1
def perfect(n):
    div = [i for i in range(1,n) if n%i==0]
    result = sum(div)
    
    if result==n:
        return 1
    else:
        return 0
```

```python
perfect(6)
```

```python
# 방법2
def perfect(n):
    for i in range(1,n+1):
        sum = 0
        for j in range(1,n):
            if i%j==0:
                sum+=i
        if sum==i:
            return 1
        else:
            return 0
```

```python
perfect(6)
```

```python
# 강의 풀이 
def perfect(n):
    s = 0
    for i in range(1,n):
        if n%i==0:
            s += i
    if s == n:
        return 1
    else:
        return 0
```

```python
perfect(5)
```

```python
for i in range(1,101):
    print(i,perfect(i))
```

```python
# 강의 풀이 2
def perfect(n):
    if sum([i for i in range(1,n) if n%i==0]) == n:
        return 1
    else:
        return 0
```

```python
for i in range(1,101):
    print(i,perfect(i))
```

```python
def perfect_print(n):
    for j in range(1,n+1):
        s = 0
        for i in range(1,j):
            if j%i==0:
                s += i
        if s == j:
            print(j)
```

```python
perfect_print(1000)
```

```python
# 과제 5-2
def rice(n):
    count = 0
    while True:
        if(n%5==0):
            count += n//5
            return count
        n = n - 3
        count+=1
```

```python
def f(n):
    c3 = 0
    c5 = 0
    while True:
        if (n%5==0):
            c5 += n//5
            return min(c3,c5)
        n -= 3
        c3 += 1
    
    if n<0:
        return -1
```

```python
f(4)
```

```python
# 강의 풀이
def rice(n):
    pode_5 = n // 5
    pode_3 = n // 3
    for i in range(pode_5,-1,-1):
        for j in range(0,pode_3+1):
            if (5*i + 3*j) == n:
                return (i+j)
    return -1
```

```python
for i in range(1,101):
    print(i,rice(i))
```

```python
def rice(n):
    pode_3 = 0
    if n%5==0:
        return n//5
    else:
        while True:
            n = n - 3
            pode_3 += 1
            if n >= 0:
                 return (pode_3 + n//5)
            else:
                return -1
```

```python
for i in range(1,101):
    print(i,rice(i))
```

```python
def rice(n):
    pode_5 = n//5
    pode_3 = n//3
    pode = [(i+j) for i in range(0,pode_5+1) for j in range(0,pode_3+1) if (5*i+3*j==n)]
    if pode:
        return min(pode)
    else:
        return -1
```
